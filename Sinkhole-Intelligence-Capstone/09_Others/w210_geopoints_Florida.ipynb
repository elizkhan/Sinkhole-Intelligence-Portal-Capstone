## Generate Points For Florida

- Latitude measures the distance north or south of the equator.
- Longitude measures distance east or west of the prime meridian.


```python
def findcounty(point, flcounty):
    for index, row in flcounty.iterrows():
        if point.within(row['geometry']):
            return (row["NAME"], row["COUNTYFP"])
```

```python
findcounty(start_point, flcounty)
```

```python
def getGeoPoint(step, fgeometry, flcounty):
    
    '''
    Step: 
    0.2 for large tile
    0.006 for small tile
    '''

    step = 2*step
    
    latmax = fgeometry.bounds[3]
    latmin = fgeometry.bounds[1]
    lonmax = fgeometry.bounds[2]
    lonmin = fgeometry.bounds[0]
#     start_point = Point(lonmin, latmax)
    start_point = Point(-87.5500, 30.9776)
    
    print("Start Point: ", lonmin, latmax)
    
    dfgpoints = pd.DataFrame()
    
    dataGP = []
    
    lat = 30.9776 #latmax
    while lat > latmin:
        lon = -87.5500  #lonmin
        while lon < lonmax:
            point = Point(lon, lat)
            if point.within(fgeometry):
                rowl = []
                rowl.append(lon)
                rowl.append(lat)
                county, countyfp = findcounty(point, flcounty)
                rowl.append(county)
                rowl.append(countyfp)
                dataGP.append(rowl)
            lon += step
        lat -= step
        
        
    dfgpoints = pd.DataFrame(dataGP, columns=["lon", "lat", "County", "CountyFP"])
    print(dfgpoints)
    
    return dfgpoints
    
```

```python
largetiles = getGeoPoint(0.1, flgeometry, flcounty)
```
