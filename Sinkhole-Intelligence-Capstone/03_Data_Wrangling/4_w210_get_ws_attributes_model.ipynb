## Get Weather Attributes from Identified Weather Station

```python
# Haversine distance formula
import numpy as np
import pandas as pd
import datetime
from datetime import datetime, timedelta

from w210_attribute_library import getwf_final, weatherFeatures
```

## Read Weather Data

```python
datadir = "../data/"
attrs = "../attrs/"
fweather = 'weather_data.csv'
dfw = pd.read_csv(datadir+fweather)
print(len(dfw))
dfw.head(1)
```

```python
len(dfw), len(dfw["stn_wban"].unique()), dfw.columns
```

```python
pivot = np.round(pd.pivot_table(dfw, values='prcp', 
                                index='year', 
                                columns='mo', 
                                aggfunc='count'),2)
pivot
```

```python
colweather = ['stn_wban', 'year', 'mo', 'da', 'temp', 'prcp', 'flag_prcp', 'min', 'max',
       'flag_prcp', 'rain_drizzle', 'sndp', 'fog', 'snow_ice_pellets', 'hail',
       'thunder', 'tornado_funnel_cloud', 'wdsp', 'mxpsd', 'gust', 'visib',
       'dewp', 'slp', 'usaf_wban', 'wban']
dfw1 = dfw[colweather]
dfw1.head(1)
```

```python
# Create a Date Field to be able to make date comparisons
dfw1["DateD"] = dfw1.apply(lambda row: datetime.strptime(str(row["year"])+"-"+str(row["mo"])+"-"+str(row["da"]),"%Y-%m-%d") ,axis=1)
dfw1.head()
```

```python
dfw1.describe()
```

```python
# Replace values with NaN
no_valid = [99.99000, 999.900000, 9999.900000, 99999.000000]
dfw1 = dfw1.replace(99.99000,np.nan)
dfw1 = dfw1.replace(999.9000000,np.nan)
dfw1 = dfw1.replace(9999.900000,np.nan)
dfw1 = dfw1.replace(99999.000000,np.nan)
```

```python
dfw1.describe()
```

```python
dfw1.head()
```

## Read Tile-Weather Station Pair Data

```python
dfF2 = pd.read_excel(datadir+ "model_tile_ws_clean.xlsx", sheet_name=0)
# dfF2["DateD"] = dfF2.apply(lambda row: datetime.strptime(row["DateD"],"%Y-%m-%d"), axis=1)
dfF2.head(1)
```

## Check Quality of Weather Station Data

```python
i = 0
wsgood = []
wsnogood = []
for index, row in dfF2.iterrows():
    wsname = row["Key_y"]
    dated = row["DateD"]
    td = timedelta(365*3)
    dftemp = dfw1[(dfw1["stn_wban"] == wsname)]
    if (dftemp["DateD"].max() > dated) & (dftemp["DateD"].min() < dated - td):
        wsgood.append(wsname)
    else:
        wsnogood.append(wsname)
```

```python
len(wsgood)
```

```python
# Should be zero for good quality stations
len(wsnogood)
```
