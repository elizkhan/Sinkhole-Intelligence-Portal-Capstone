## Get Weather Attributes from Identified Weather Station

```python
# Haversine distance formula
import numpy as np
import pandas as pd
import datetime
from datetime import datetime, timedelta

from w210_attribute_library import getwf_final, weatherFeatures
```

## Read Weather Data

```python
datadir = "../data/"
attrs = "../attrs/"
fweather = 'weather_data.csv'
dfw = pd.read_csv(datadir+fweather)
print(len(dfw))
```

```python
len(list(dfw["stn_wban"].unique()))
```

```python
pivot = np.round(pd.pivot_table(dfw, values='prcp', 
                                index='year', 
                                columns='mo', 
                                aggfunc='count'),2)
pivot
```

```python
colweather = ['stn_wban', 'year', 'mo', 'da', 'temp', 'prcp', 'flag_prcp', 'min', 'max',
       'flag_prcp', 'rain_drizzle', 'sndp', 'fog', 'snow_ice_pellets', 'hail',
       'thunder', 'tornado_funnel_cloud', 'wdsp', 'mxpsd', 'gust', 'visib',
       'dewp', 'slp', 'usaf_wban', 'wban']
dfw1 = dfw[colweather]
dfw1.head(1)
```

```python
# Create a Date Field to be able to make date comparisons
dfw1["DateD"] = dfw1.apply(lambda row: datetime.strptime(str(row["year"])+"-"+str(row["mo"])+"-"+str(row["da"]),"%Y-%m-%d") ,axis=1)
```

```python
dfw1["DateD"].min()
```

```python
dfw1.describe()
```

```python
# Replace values with NaN
no_valid = [99.99000, 999.900000, 9999.900000, 99999.000000]
dfw1 = dfw1.replace(99.99000,np.nan)
dfw1 = dfw1.replace(999.9000000,np.nan)
dfw1 = dfw1.replace(9999.900000,np.nan)
dfw1 = dfw1.replace(99999.000000,np.nan)
```

```python
dfw1.describe()
```

```python
dfw1.head()
```

## Read Tile-Weather Station Pair Data

```python
df365 = pd.read_csv(attrs+ "model_wspair_sat_sh_soil_attr_365.csv")
df365["DateD"] = df365.apply(lambda row: datetime.strptime(row["DateD"],"%Y-%m-%d"), axis=1)
```

## Get Weather Features

### Get Weather Attributes with timedelta = 365

```python
# date_string = '2022/07/05'

date_string = 'none'
daysdelta = timedelta(365)
is_scale = False

dfwf = weatherFeatures(df365, dfw1, is_scale, daysdelta, date_string)

dfwf.to_csv(attrs + "model_ws_attr_365.csv", index=False)
```
