At the moment bnlearn can only be used for discrete/categorical modeling. There are possibilities to model your data though. You can for example discretize your variables with domain/experts knowledge or maybe a more data-driven threshold. Lets say, if you have a temperature, you can mark temperature < 0 as freezing, and >0 as normal. Or many smaller categories.

```python
import bnlearn as bn
import numpy as np
import pandas as pd
import datetime
from datetime import datetime
import os
import matplotlib.pyplot as plt
import math
import random

from sklearn.model_selection import train_test_split

dirname = '../modeldata/'
dirpm = '../prediction_train_test/'
```

```python
finEvents = pd.read_csv(dirname+"data_model1_365.csv")
finEvents.isnull().values.any(), finEvents.isnull().sum().sum()
```

```python
finEvents = finEvents.dropna()
finEvents.isnull().values.any(), finEvents.isnull().sum().sum()
```

```python
finEvents.columns
```

```python
takeout = set(['rolling_7_precip', 'y2_mean_tmp', 'rolling_7_precip'])
dfvars = pd.read_csv(dirname+"attr80.csv")
x_variables_initial = list(dfvars["attribute"].unique())
```

```python
setv = set(x_variables_initial)
setv = setv - takeout
x_variables = list(setv)
```

```python
print(len(x_variables))
features = finEvents[['Sinkhole']+x_variables]
print(x_variables)
```

```python
means = features[x_variables].mean()
list(means[means.index == 'l200'])[0]
```

```python
newlist = ['Sinkhole']
featurelist = []
for c in x_variables:
    features[c+"_b"] = features.apply(lambda r: 1 if (r[c]> list(means[means.index == c])[0]) else 0, axis=1)
    newlist.append(c+"_b")
    featurelist.append(c+"_b")
```

```python
features2 = features[newlist]
features2.head()
```

```python
train, test = train_test_split(features2, test_size=0.2)
```

```python
len(train)
```

```python
# Learn the DAG in data using Bayesian structure learning:
DAG = bn.structure_learning.fit(train)
```

```python
# print adjacency matrix
DAG['adjmat']
```

```python
# Plot
G = bn.plot(DAG)

# Interactive plotting
G = bn.plot(DAG, interactive=True)
```

```python
model = bn.parameter_learning.fit(DAG, features2, methodtype='bayes')
```

```python
q1 = bn.inference.fit(model, variables=featurelist, evidence={'Sinkhole':0})
print(q1.features2)
```
